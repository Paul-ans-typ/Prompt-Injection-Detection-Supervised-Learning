```python
# Cell 1: Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
# Cell 2: Load Data
df = pd.read_csv('../data/raw/MPDD.csv')
```

```python
# Cell 3: Check for Missing Values
print("Missing values:\n", df.isnull().sum())
```

```python
# Cell 4: Class Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='isMalicious')
plt.title('Distribution of Benign vs. Malicious Prompts')
plt.show()

print("Class balance:\n", df['isMalicious'].value_counts(normalize=True))
```

```python
# Cell 5: Text Length Analysis
df['word_count'] = df['Prompt'].apply(lambda x: len(str(x).split()))

print("Word Count Statistics:\n", df['word_count'].describe())

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='word_count', hue='isMalicious', bins=50, kde=True)
plt.title('Word Count Distribution by Class')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.xlim(0, df['word_count'].quantile(0.99)) 
plt.show()
```
